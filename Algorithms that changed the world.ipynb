## Quickhull algorithum 

Complexity

Worst case time complexity: Θ(N^2) <br></br>
Average case time complexity: Θ(N log N) <br></br>
Best case time complexity: Θ(N log N) <br></br>
Space complexity: Θ(N)

Making random points on a plot

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.random.uniform(size=(2,30))

print(x, '\n')
print(y)


for i in range(0, len(x)):
    plt.plot(x[i],y[i], 'ro')

plt.show()
```

Connecting two pionts

```python
plt.plot(x,y, 'ro')

def connectpoints(p1,p2):
    print(p1)
    print(p2)
    x1, x2 = p1[0], p2[0]
    y1, y2 = p1[1], p2[1]

    plt.plot([x1,x2], [y1,y2],'k-')
    


    
current_point = [x[0],y[0]]

destination_point = [x[5],y[5]]



connectpoints(current_point,destination_point)


plt.show()
```

Divide the points into two subsets

```python
plt.plot(x,y, 'ro')

def sort(x,y):
    array = []
    for i in range(0, len(x)):
        temp = []
        temp.append(x[i])
        temp.append(y[i])
        array.append(temp)
    return array
    


def distance_points(points):
    left_p = [1,1]
    right_p = [0,0]
    for p in points:
        if p[0] < left_p[0]:
            left_p = p
            
        if p[0] > right_p[0]:
            right_p = p
        
    return left_p, right_p
        
        
    

points = sort(x,y)


leftM_point, rightM_p = distance_points(points)

connectpoints(leftM_point, rightM_p)



    
```

We have the leftmost and right most pionts, and we can split each half into two subsets: above and below the line.
Now we find the points to connect to around each subset
    

```python
#Define the two subsets

isabove = lambda p, a,b: np.cross(p-a, b-a) < 0

above_arr = []
below_arr = []
a = np.array(leftM_point)
b = np.array(rightM_p)

for i in points:
    p1 = np.array(i)
   
    if p1[0] != a[0] and p1[0] != b[0]:
        if isabove(p1, a, b):
            above_arr.append(i)
        else:
            below_arr.append(i)
    




data1 = np.array(above_arr)
x, y = data1.T
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.plot(x,y, 'ro', label = "1")

data2 = np.array(below_arr)
x, y = data2.T
plt.plot(x,y, 'bo', label= "2")

x, y = a.T    
plt.plot(x, y, 'go')    

x, y = b.T    
plt.plot(x, y, 'go') 

connectpoints(leftM_point,rightM_p)
```
